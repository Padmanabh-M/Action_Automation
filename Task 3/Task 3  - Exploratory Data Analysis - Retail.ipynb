# Making additional column that shows if either a profit was made or a loss was incurred 

```python
def proloss(data):
    if data < 0:
        return "Loss"
    else:
        return "Profit"
    
df['Proloss'] = df['Profit'].apply(proloss)
df
```

```python
df.columns = ['ShipMode', 'Segment', 'Country', 'City', 'State', 'PostalCode',
       'Region', 'Category', 'SubCategory', 'Sales', 'Quantity', 'Discount',
       'Profit','Proloss']
```

```python
dis = df['Profit'][df['Proloss'] == 'Profit'].sum()
print('Profit : ', dis)
dis1 = df['Profit'][df['Proloss'] == 'Loss'].sum()
print('Loss : ',dis1)
print('\n')

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
What = ['Profit', 'Loss']
Val = [dis, dis1*-1]
ax.bar(What,Val)
plt.ylabel('Values in USD', size = 18)
plt.show()
```

```python
def Analyse_State(Data):
    first = df.groupby(['State'])
    for i in Data:
        data = first.get_group(i).groupby(['Category'])
        fig, ax = plt.subplots(1, 3, figsize = (30,6))
        fig.suptitle(i, fontsize=18)        
        ax_index = 0
        for ii in ['Furniture', 'Office Supplies', 'Technology']:
            ii_data = data.get_group(ii).groupby(['SubCategory']).sum()
            sns.barplot(x = ii_data.Profit, y = ii_data.index, ax = ax[ax_index])
            ax[ax_index].set_ylabel(i)
            ax_index +=1
        fig.show()
```
